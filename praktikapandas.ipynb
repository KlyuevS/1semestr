```python
import pandas as pd #задание 1 первая таблица
unames = ('user_id','gender','age','occupation','zip')
users = pd.read_table('users.dat', sep='::', engine='python', names = unames)
users

```

```python
rnames=['user_id', 'movie_id', 'rating', 'timestamp'] #ыторая таблица
ratings=pd.read_table('ratings.dat', sep='::', engine='python', names = rnames)
ratings
```

```python
nnames=['movie_id','title','genres'] #третья таблица
movies=pd.read_table('movies.dat', sep='::', engine='python', names = nnames)
movies
```

```python
data=pd.merge(ratings,users)  #задание 2
data
```

```python
data=pd.merge(data,movies)
data
```

```python
mean_ratings_gender=data.pivot_table('rating', index='age',columns='gender', aggfunc='mean') # задание 3
mean_ratings_gender
```

```python
mean_ratings_age=data.pivot_table('rating', index='title',columns='age', aggfunc='mean')
mean_ratings_age
```

```python
sortage = mean_ratings_gender.sort_values(by='age',
                 ascending=False)#задание 4
sortage
```
